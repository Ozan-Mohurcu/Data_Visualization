# Coklu Data ve Legend Verme


```python
age = np.arange(25, 44, 2.5)

salary_1 = np.arange(40000, 76000, 5000)

salary_2 = [45372, 46852, 69401, 56790, 73000, 80000, 83545, 95653]
```

```python
plt.plot(age, salary_1, "b:", label="Salary_1")

plt.plot(age, salary_2, "r-", label="Salary_2")

plt.xlabel("Age")

plt.ylabel("Salary")

plt.title("SALARY BY AGE")

plt.legend() # legend dedik.

plt.show()
```

```python
plt.plot(age, salary_1, "b:", label="Salary_1")

plt.plot(age, salary_2, "r-", label="Salary_2")

plt.xlabel("Age")

plt.ylabel("Salary")

plt.title("SALARY BY AGE")

plt.legend(loc='upper center') # default sol üst, etiketin konumunu degistirdik.

plt.show()
```

```python
plt.plot(age, salary_1, "g:", label="Salary_1")

plt.plot(age, salary_2, "m-", label="Salary_2")

plt.xlabel("Age")

plt.ylabel("Salary")

plt.title("SALARY BY AGE")

plt.legend(bbox_to_anchor=(0.6, 0.8)) #İnce ayar yaptik: (0.6 (x), 1(y))

plt.show()
```

```python
# line kalinligi

plt.figure(figsize=(10, 6))

plt.plot(age, salary_1, "b-", linewidth=6, label="Salary 1")

plt.plot(age, salary_2, "r-.", lw=10, label="Salary 2")

plt.legend()

plt.show()
```

# Figure ve Axes Renklendirme

```python
plt.figure(facecolor="yellow", figsize=(10, 5)) # cerceve rengi

ax = plt.axes()

ax.set_facecolor("xkcd:jade") # grafik rengi, xkcd'yi girmemiz gerekli katalogdan secim yapiyorsak: https://matplotlib.org/stable/users/explain/colors/colors.html

plt.plot(age, salary_1, "b-", linewidth=6, label="Salary 1")

plt.plot(age, salary_2, "r-.", lw=2, label="Salary 2")

plt.xlabel("Age")

plt.ylabel("Salary")

plt.title("Salary by Age")

plt.legend(bbox_to_anchor=(0, 0.5, 0.57, 0.5)) # Ilk iki deger x ve y eksenine göre kaydirir.
                                               # son iki tanesi ise width ve height'a göre kaydirir.
                                               
plt.show()
```

# Function Method and Object Oriented Methods
### Functional Method

```python
plt.plot(age, salary_1)

plt.xlabel("Age")

plt.ylabel("Salary")

plt.title("Salary by Age")

plt.show()
```

# Object Oriented Method

```python
fig, ax = plt.subplots() # Genel kalip

ax.plot(age, salary_1, "b")

ax.set_xlabel("age")

ax.set_ylabel("Salary")

ax.set_title("Salary by Age");
```

# Functional Method (Coklu Grafik Cizimi)

```python
plt.subplot(2, 1, 1) # 2 satirlik 1 sütunluk figür'ün 1. grafigi

plt.plot(age, salary_1, "r")

plt.subplot(2, 1, 2) # 2 satirlik 1 sütunluk figür'ün 2. grafigi

plt.plot(age, salary_2, "b");
```

```python
plt.subplot(211) # Bu sekilde de yazabiliriz.

plt.plot(age, salary_1, "r")

plt.subplot(212)

plt.plot(age, salary_2, "b");
```

```python
# Tüm datayi ayni grafikte verir. Cünkü konum bilgisi girmedim.

x = np.arange(0, 30 , 0.2)

fig, ax = plt.subplots(figsize=(14, 8))

ax.plot(x, x-1, color = "r", linewidth=1, linestyle="-")

ax.plot(x, x-2, color = "green", lw=3, ls="-.")

ax.plot(x, x-3, color = "m", lw=3, ls=":")

ax.plot(x, x-4, color = "cyan", lw=5, ls="--");
```

```python
x = np.arange(0, 10)

fig, ax = plt.subplots(2, 2, figsize =(12, 6))

plt.suptitle("LineWidth & Marker & MarkerSize & Color")

ax[0, 0].plot(x, "r", marker="+", markersize="5") # Konum girdigimiz icin 4 ayri grafik geldi.

ax[0, 1].plot(x-2, "y", marker="o", markersize="6")

ax[1, 0].plot(x-3, "black", marker="s", markersize="7", lw=0)

ax[1, 1].plot(x-4, "r", marker = "1", ms="8");
```

```python
x = np.arange(0, 10)

fig, ax = plt.subplots(2, 2, figsize =(12, 6))

plt.suptitle("LineWidth & Marker & MarkerSize & Color", fontsize = 20, y = 1)

ax[0,0].plot(x, "r", marker = "+", markersize="5")

ax[0,0].set_title("Grafik-1", fontsize=18)

ax[0,0].set_xlabel("X", fontsize=18)

ax[0,0].set_ylabel("Y")

ax[0,1].plot(x-2, "y", marker = "o", markersize="5")

ax[0,1].set_title("Grafik-2", fontsize=18)

ax[0,1].set_xlabel("X", fontsize=18)

ax[0,1].set_ylabel("Y", fontsize=18)

ax[1,0].plot(x-3, "black", marker = "s", markersize="5", lw=0)

ax[1,0].set_title("Grafik-3", fontsize=18)

ax[1,0].set_xlabel("X", fontsize=18)

ax[1,0].set_ylabel("Y", fontsize=18)

ax[1,1].plot(x-4, "r", marker = "1", ms="5")

ax[1,1].set_title("Grafik-4", fontsize=18)

ax[1,1].set_xlabel("X", fontsize=18)

ax[1,1].set_ylabel("Y", fontsize=18)

plt.tight_layout();
```

```python
x = np.arange(-50, 50)

y1 = np.power(x, 2)

y2 = np.power(x, 3)

y3 = np.sin(x)

y4 = np.cos(x)

y5 = np.tan(x)

y6 = np.tanh(x)

y7 = np.sinh(x)

y8 = np.cosh(x)

y9 = np.exp(x)

fig1, ax1 = plt.subplots(nrows=3, ncols=3, figsize=(20, 20))

ax1[0, 0].plot(x, y1, "tab:blue") # set the color of line chart

ax1[0, 0].set_title("Square Function") # setting title of subplot

ax1[0, 0].set_xlabel("X", fontsize=18)

ax1[0, 0].set_ylabel("Y", fontsize=18)

ax1[0, 1].plot(x, y2, "tab:orange") # set the color of line chart

ax1[0, 1].set_title("Cubic Function") # setting title of subplot

ax1[0, 1].set_xlabel("X", fontsize=18)

ax1[0, 1].set_ylabel("Y", fontsize=18)

ax1[0, 2].plot(x, y3, "tab:green") # set the color of line chart

ax1[0, 2].set_title("Sine Function") # setting title of subplot

ax1[0, 2].set_xlabel("X", fontsize=18)

ax1[0, 2].set_ylabel("Y", fontsize=18)

ax1[1, 0].plot(x, y4, "r-") # set the color of line chart

ax1[1, 0].set_title("Cosine Function") # setting title of subplot

ax1[1, 0].set_xlabel("X", fontsize=18)

ax1[1, 0].set_ylabel("Y", fontsize=18)

ax1[1, 1].plot(x, y5, "b-") # set the color of line chart

ax1[1, 1].set_title("Tangent Function") # setting title of subplot

ax1[1, 1].set_xlabel("X", fontsize=18)

ax1[1, 1].set_ylabel("Y", fontsize=18)

ax1[1, 2].plot(x, y6, "g-") # set the color of line chart

ax1[1, 2].set_title("Hyperbolic Tangent Function") # setting title of subplot

ax1[1, 2].set_xlabel("X", fontsize=18)

ax1[1, 2].set_ylabel("Y", fontsize=18)

ax1[2, 0].plot(x, y7, "m-") # set the color of line chart

ax1[2, 0].set_title("Hyperbolic Sine Function") # setting title of subplot

ax1[2, 0].set_xlabel("X", fontsize=18)

ax1[2, 0].set_ylabel("Y", fontsize=18)

ax1[2, 1].plot(x, y8, "y*") # set the color of line chart

ax1[2, 1].set_title("Hyperbolic Cosine Function") # setting title of subplot

ax1[2, 1].set_xlabel("X", fontsize=18)

ax1[2, 1].set_ylabel("Y", fontsize=18)

ax1[2, 2].plot(x, y9, "kv") # set the color of line chart

ax1[2, 2].set_title("Exp Function") # setting title of subplot

ax1[2, 2].set_xlabel("X", fontsize=18)

ax1[2, 2].set_ylabel("Y", fontsize=18)

plt.show()
```
