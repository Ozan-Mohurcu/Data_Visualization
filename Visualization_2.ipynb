```python
import matplotlib.pyplot as plt
import numpy as np
```

# 1. Çizgi Grafiği

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure()
plt.plot(x, y, label='sin(x)', color='blue')
plt.title('Çizgi Grafiği')
plt.xlabel('X Değeri')
plt.ylabel('Y Değeri')
plt.legend()
plt.show()
```

# 2. Sütun Grafiği

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
plt.figure()
plt.bar(categories, values, color='green')
plt.title('Sütun Grafiği')
plt.xlabel('Kategoriler')
plt.ylabel('Değerler')
plt.show()
```

# 3. Pasta Grafiği

```python
labels = ['Elma', 'Armut', 'Kiraz', 'Portakal']
sizes = [20, 30, 25, 25]
plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pasta Grafiği')
plt.show()
```

# 4. Histogram


```python
data = np.random.randn(1000)
plt.figure()
plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Değerler')
plt.ylabel('Frekans')
plt.show()
```

# 5. Dağılım Grafiği

```python
x = np.random.rand(100)
y = np.random.rand(100)
plt.figure()
plt.scatter(x, y, color='red', alpha=0.5)
plt.title('Dağılım Grafiği')
plt.xlabel('X Değeri')
plt.ylabel('Y Değeri')
plt.show()
```

# 6. Boxplot Grafiği

```python
data = [np.random.randn(100) for _ in range(4)]
plt.figure()
plt.boxplot(data)
plt.title('Kutu Grafiği')
plt.xlabel('Kategoriler')
plt.ylabel('Değerler')
plt.show()
```

# 7. Alan Grafiği

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x) + 0.5
plt.figure()
plt.fill_between(x, y1, y2, color='lightblue', alpha=0.5)
plt.title('Alan Grafiği')
plt.xlabel('X Değeri')
plt.ylabel('Y Değeri')
plt.show()
```

# 8. Hata Çubuklu Grafiği

```python
x = np.arange(0, 5, 1)
y = x**2
errors = [0.5, 0.4, 0.3, 0.2, 0.1]
plt.figure()
plt.errorbar(x, y, yerr=errors, fmt='o', ecolor='black', capsize=5, label='Veriler')
plt.title('Hata Çubuklu Grafiği')
plt.xlabel('X Değeri')
plt.ylabel('Y Değeri')
plt.legend()
plt.show()
```

# 9. Çubuk Yığılmalı Grafiği

```python
labels = ['Grup 1', 'Grup 2', 'Grup 3']
men_means = [20, 35, 30]
women_means = [25, 32, 34]
x = np.arange(len(labels))
width = 0.35
plt.figure()
plt.bar(x - width/2, men_means, width, label='Erkek')
plt.bar(x + width/2, women_means, width, label='Kadın')
plt.title('Çubuk Yığılmalı Grafiği')
plt.xlabel('Gruplar')
plt.ylabel('Değerler')
plt.xticks(x, labels)
plt.legend()
plt.show()
```

# 10. Isı Haritası

```python
data = np.random.rand(10, 10)
plt.figure()
plt.imshow(data, cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.title('Isı Haritası')
plt.show()
```
